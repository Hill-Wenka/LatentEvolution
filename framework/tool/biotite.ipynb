## convert atom array to sequence and compare to the fasta sequence

```python
sequences, descriptions = read_fasta(fasta_path)
sequences, descriptions
```

```python
convert_seq = atom_array_to_sequence(atom_array)
convert_seq, len(convert_seq)
```

```python
sequences[0], len(sequences[0])
```

```python
for i in range(len(list(sequences[0]))):
    if i >= len(convert_seq):
        print(i, sequences[0][i], 'None')
    else:
        if sequences[0][i] == convert_seq[i]:
            pass
        else:
            print(i, sequences[0][i], convert_seq[i])
```
